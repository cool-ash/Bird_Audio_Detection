### **Testing Dataset**

### **Model**

```python
model = models.Sequential()
model.add(layers.InputLayer(input_shape=(862, 40, 1)))
model.add(layers.Conv2D(8, (7,7), activation='tanh', padding='same'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2), padding='same'))
model.add(layers.Conv2D(16, (5,5), activation='tanh', padding='same'))
model.add(layers.MaxPooling2D((2, 2), padding='same'))
model.add(layers.Dropout(0.4))
model.add(layers.Conv2D(32, (5,5), activation='tanh', padding='same'))
model.add(layers.MaxPooling2D((2, 2), padding='same'))
model.add(layers.Dropout(0.4))
model.add(layers.Conv2D(32, (5,5), activation='tanh', padding='same'))
model.add(layers.MaxPooling2D((2, 2), padding='same'))
model.add(layers.Dropout(0.4))
model.add(layers.Conv2D(64, (5,5), activation='tanh', padding='same', kernel_regularizer=l1_l2(l1=0.001, l2=0.001)))
model.add(layers.MaxPooling2D((2, 2), padding='same'))
model.add(layers.Dropout(0.5))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='tanh', kernel_regularizer=l2(0.001)))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.summary()
```
